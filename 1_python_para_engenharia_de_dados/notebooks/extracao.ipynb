```python
import pandas as pd
import requests
import json
import csv
```

```python
path_json = r'../data_raw/dados_empresaA.json'
path_csv = r'../data_raw/dados_empresaB.csv'
```

```python
with open(path_json, 'r') as file:
    data_empresaA = json.load(file)

print(data_empresaA)
```

```python
dados_empresaB = []

with open(path_csv, 'r') as file:
    spamreader = csv.DictReader(file, delimiter=",")

    for i, row in enumerate(spamreader):
        dados_empresaB.append(row)
```

```python
# alterando keys
print(list(data_empresaA[0].keys()))
```

```python
print(list(dados_empresaB[0].keys()))
```

```python
mapping_keys_empresaA = {
    #    B                A
    'Nome do Item': 'Nome do Produto',
    'Classificação do Produto': 'Categoria do Produto',
    'Valor em Reais (R$)': 'Preço do Produto (R$)',
    'Quantidade em Estoque': 'Quantidade em Estoque',
    'Nome da Loja': 'Filial',
    'Data da Venda': 'Data da Venda'
}

for data in dados_empresaB:
    temp_dict = {}
    
    for key in list(data.keys()):
        if key in mapping_keys_empresaA: 
            temp_dict[mapping_keys_empresaA[key]] = data.pop(key)
    
    data.update(temp_dict)

for d in dados_empresaB:
    d.update({mapping_keys_empresaA.get(k, k): d.pop(k) for k in list(d.keys()) if k in mapping_keys_empresaA})
```

```python
dados_empresaB
```

```python
datas = data_empresaA + dados_empresaB
cols_default = list(dados_empresaB[0].keys())
```

```python
data_tratado = []

for d in datas:
    dict = {}

    for col in cols_default:
        dict[col] = d.get(col, 'Indisponível')
    
    data_tratado.append(dict)
```

```python
for row in data_tratado:
    print(row)
```

```python
path_processed_data = '../data_processed/data_consolidado.csv'

with open(path_processed_data, 'w') as file:
    writer = csv.DictWriter(file, fieldnames=list(data_tratado[0].keys()))
    writer.writeheader()

    for row in data_tratado:
        writer.writerow(row)
```

```python
cols = list(data_tratado[0].keys())
list_values = [cols]
list_values += [[d[col] for col in cols] for d in data_tratado]
```
